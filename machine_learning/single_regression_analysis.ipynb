```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
import matplotlib.pyplot as plt
import seaborn

df = pd.read_csv('time.csv', header=0)
data = list()

for column in df.columns:
  X = range(0, 5)
  X = np.array(X).reshape(-1,1)
  y = df[column]

  model = LinearRegression()
  model.fit(X, y)

  plt.plot(X, y, 'o')
  plt.plot(X, model.predict(X), linestyle="solid")
  plt.show()

  print('モデル関数の回帰変数 w1: %.3f' %model.coef_)
  print('モデル関数の切片 w2: %.3f' %model.intercept_)
  # print('y= %.3fx + %.3f' % (model.coef_ , model.intercept_))
  # print('決定係数 R^2： ', model.score(X, y))
  data.append(model.coef_*5 + model.intercept_)
```

```python
import csv

csv_path = r"test.csv"

with open(csv_path, 'w') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```
